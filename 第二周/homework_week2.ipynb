```python
def add_padding(img, kernel_size):
    p = kernel_size // 2
    black = [0,0,0]
    padding_img = cv2.copyMakeBorder(img, p,p,p,p, cv2.BORDER_CONSTANT, value=black)
    return padding_img

def convolution_x(img):
    kernel = np.array([-1,0,1])
    img_h, img_w = img.shape
    board = 1
    new_image = np.zeros((6,6), np.uint8)
    
    for i in range(board, img_h-board):
        for j in range(board, img_w-board):
            new_image[i-1][j-1] = np.sum(kernel * img[i, j-board:j+board+1])
    new_image = new_image.astype(np.uint8)
    return new_image/255

def convolution_y(img):
    kernel = np.transpose(np.array([-1,0,1]))
    
    img_h, img_w = img.shape
    board = 1
    new_image = np.zeros((6,6), np.uint8)
    for i in range(board, img_h-board):
        for j in range(board, img_w-board):
            new_image[i-1][j-1] = np.sum(kernel * img[i-board:i+board+1, j])
    new_image = new_image.astype(np.uint8)
    return new_image/255

def calcTangent(dx, dy):
    res = np.zeros(dx.shape, np.float)
    for i in range(6):
        for j in range(6):
            if (dx[i][j] == 0):
                if (dy[i][j] >= 0):
                    res[i][j] = 90
                else:
                    res[i][j] = 270
            else:
                res[i][j] = np.arctan(dy[i][j] / dx[i][j])
    res[res >= 90] = 1            
    return res.astype(np.uint8).flatten()
    

def calcMagnitude(dx, dy):
    res = np.sqrt(dx**2 + dy**2)
    res[res < 0.1] = 0
    return res


image_y_0 = convolution_y(add_padding(num_0.numpy().copy(),3))
image_x_0 = convolution_x(add_padding(num_0.numpy().copy(),3))
# print(image_y)
# print(image_x)

image_y_9 =  convolution_y(add_padding(num_9.numpy().copy(),3))
image_x_9 = convolution_x(add_padding(num_9.numpy().copy(),3))

image_y_9_cp =  convolution_y(add_padding(num_9_cp.numpy().copy(),3))
image_x_9_cp = convolution_x(add_padding(num_9_cp.numpy().copy(),3))

image_y_4 =  convolution_y(add_padding(num_4.numpy().copy(),3))
image_x_4 = convolution_x(add_padding(num_4.numpy().copy(),3))


# print(calcTangent(image_x,image_y))
# print(calcTangent(image_x_9_cp,image_y_9_cp))
# print("\n")
# print(calcTangent(image_x_9,image_y_9))
# print("\n")
# print(calcTangent(image_x_0,image_y_0))
# print("\n")
# print(calcTangent(image_x_4,image_y_4))
# print(calcValue(image_x_0,image_y_0))
# print(calcMagnitude(image_x_9,image_y_9))
# print(calcMagnitude(image_x_9_cp,image_y_9_cp))
# print(calcHamming(calcValue(image_x_9,image_y_9),calcValue(image_x_9_cp,image_y_9_cp).flatten()))
print(calcHamming(calcTangent(image_x_9,image_y_9), calcTangent(image_x_9_cp,image_y_9_cp)))

# print(image_x)
# print(image_y)
```

```python
def model_calcDistance(feature):
    # 下面添加对feature进行决策的代码，判定出feature 属于[0,1,2,3,...9]哪个类别
    scores = []
    for i in range(10):
        scores.append(calcHamming(feature, features[i]))
    y = np.argmin(scores)
    return y

# def
```
